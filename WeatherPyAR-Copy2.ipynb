```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import openweathermapy.core as owm
import seaborn as sns

# Import API key
from api_keys import api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

```

## Generate Cities List

```python
cities=[]
for x in range(600):
    lat, lng = np.random.uniform(low=-90.000, high=90.000), np.random.uniform(low=-180.000, high=180.000)
    city = citipy.nearest_city(lat, lng)
    
    if city not in cities:
        cities.append(city.city_name)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
#Temperature
#Humidity
#Cloudiness
#Wind Speed
```

```python
#Test to see the location information 
base_url = "http://api.openweathermap.org/data/2.5/weather?"   
target_city = "Paris"
target_url = base_url + "appid=" + api_key + "&units=IMPERIAL" + "&q=" + target_city.replace(" ","+")
target_url
```

```python
#Create columns for dataframe to store the information 
City=[]
Country=[]
ID=[]
Temperature=[]
Latitude=[]
Longitude=[]
Humidity=[]
Cloudiness=[]
WindSpeed=[]

# Loop through the list of cities and add the data to the df.
base_url = "http://api.openweathermap.org/data/2.5/weather?"

for city in cities:
    target_city = city
    target_url = base_url + "appid=" + api_key + "&units=IMPERIAL" + "&q=" + target_city.replace(" ","+")
    city_data_info = requests.get(target_url).json()
   
    if city_data_info["cod"] == "404":
        print("City not found, skipping...")
    else:
        City.append(city_data_info["name"])
        Country.append(city_data_info["sys"]["country"])
        ID.append(city_data_info["id"])
        Temperature.append(city_data_info["main"]["temp"])
        Latitude.append(city_data_info["coord"]["lat"])
        Longitude.append(city_data_info["coord"]["lon"])
        Humidity.append(city_data_info["main"]["humidity"])
        Cloudiness.append(city_data_info["clouds"]["all"])
        WindSpeed.append(city_data_info["wind"]["speed"])
        
        
        print("Proceesing: City:" , city_data_info["name"], ",ID:", city_data_info["id"])
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
#Convert the information to a dataframe
data_city = {
    "Name": City,
    "Country": Country,
    "ID": ID,
    "Cloudiness (%)": Cloudiness,
    "Temperature": Temperature,
    "Latitude": Latitude,
    "Humidity (%)": Humidity,
    "Wind Speed (mph)": WindSpeed
}
#Convert to dataframe and label columns
city_weather_df = pd.DataFrame(data_city, columns=("Name","Country", "ID", "Temperature","Latitude","Humidity (%)","Cloudiness (%)","Wind Speed (mph)"))
city_weather_df.head()
```

```python
# Save weather data to csv file
city_weather_df.to_csv("cityweatherdata.csv", encoding="utf-8", index=False)
```
